```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
timings = np.load('timings.npy')
```

```python
n = 100
```

```python
# Range
epochs = np.arange(1,len(timings)+1,1, dtype=int)
```

```python
# Get the average epoch length evert 100 epochs
timings_averaged = np.mean(timings.reshape(-1, n), axis=1)
epochs_average = np.arange(n, len(timings)+1, n, dtype=int)
```

```python
plt.plot(epochs_average, timings_averaged)
plt.xlabel("Epoch")
plt.ylabel("Time (Seconds)")
plt.show()
```
